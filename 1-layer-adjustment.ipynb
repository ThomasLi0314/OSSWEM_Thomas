```python
import matplotlib.pyplot as plt
import numpy as np
from OSSWEM import SSWEM
```

```python
# Geostrophic adjustment problem
M = SSWEM(80,
          10., # Gravity [m s-2]
          10., # Max depth [m]
          2000.e3, # Domain width [m]
          1.e-4, # Coriolis [s-1]
          0.e-11, # df/dy [m-1 s-1]
          0.e-3, # Drag rate [m-1 s-1]
          0e3 # Viscosity [m s-2]
          )
M.perturb_eta(0.01, 200.e3, 1000e3, y0=1000e3)
plt.pcolormesh( M.xq/1e3, M.yq/1e3, M.eta(k=0) ); plt.colorbar();

u,v,h,time = M.run(50.,1,1250);
plt.figure();
plt.contourf( M.xh1, M.yh1, M.eta(h[-1], k=0) ); plt.colorbar();
```
